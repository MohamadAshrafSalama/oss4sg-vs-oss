```python
def remove_outliers(df, column_list):
    for column in column_list:
        if column in df.columns:
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
import pandas as pd
import os
from collections import Counter
import re

# Define the directories for oss and oss4sg projects
base_dir = '/Users/mohamadashraf/Desktop/Projects/OSS4SG_Classification-main/Datasets'
oss_dir = os.path.join(base_dir, 'oss_projects')
oss4sg_dir = os.path.join(base_dir, 'oss4sg_projects')

# Initialize lists to collect contributor names
oss_contributors = []
oss4sg_contributors = []

# Function to check if a string is purely numeric
def is_not_numeric(s):
    return not s.isdigit()

# Function to process each project directory
def process_contributors(directory, contributor_list):
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            try:
                # Read the CSV file without headers
                df = pd.read_csv(file_path, header=None)
                # Use the first column as the 'User' column
                users = df.iloc[:, 0].astype(str).dropna().tolist()
                # Filter out numeric names, 'Total Users', and 'user'
                filtered_users = [
                    user for user in users
                    if is_not_numeric(user)
                    and not re.match(r'Total\s+Users', user, re.IGNORECASE)
                    and user.lower() != 'user'
                ]
                contributor_list.extend(filtered_users)
            except pd.errors.EmptyDataError:
                print(f"No data in file: {filename}")

# Process OSS projects
process_contributors(oss_dir, oss_contributors)

# Process OSS4SG projects
process_contributors(oss4sg_dir, oss4sg_contributors)

# Count the occurrences of each contributor in OSS and OSS4SG
oss_contributors_count = Counter(oss_contributors)
oss4sg_contributors_count = Counter(oss4sg_contributors)

# Convert these counts to a DataFrame and save to CSV
oss_contributors_df = pd.DataFrame(
    list(oss_contributors_count.items()),
    columns=['Contributor', 'Projects Count']
)
oss4sg_contributors_df = pd.DataFrame(
    list(oss4sg_contributors_count.items()),
    columns=['Contributor', 'Projects Count']
)

oss_contributors_df.to_csv(
    os.path.join(base_dir, 'oss_contributors_count.csv'),
    index=False
)
oss4sg_contributors_df.to_csv(
    os.path.join(base_dir, 'oss4sg_contributors_count.csv'),
    index=False
)
```

```python
import pandas as pd
import os
from collections import Counter
import re
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn2

# Define the directories for OSS and OSS4SG projects
base_dir = '/Users/mohamadashraf/Desktop/Projects/OSS4SG_Classification-main/Datasets'
oss_dir = os.path.join(base_dir, 'oss_projects')
oss4sg_dir = os.path.join(base_dir, 'oss4sg_projects')

# Function to check if a string is purely numeric
def is_not_numeric(s):
    return not s.isdigit()

# Function to process each project directory
def process_contributors(directory):
    project_contributors = {}
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            project_name = os.path.splitext(filename)[0]
            try:
                # Read the CSV file without headers
                df = pd.read_csv(file_path, header=None)
                # Use the first column as the 'User' column
                users = df.iloc[:, 0].astype(str).dropna().tolist()
                # Filter out numeric names, 'Total Users', and 'user'
                filtered_users = [
                    user.strip() for user in users
                    if is_not_numeric(user)
                    and not re.match(r'Total\s+Users', user, re.IGNORECASE)
                    and user.lower() != 'user'
                ]
                # Assign contributors to the project
                project_contributors[project_name] = filtered_users
            except pd.errors.EmptyDataError:
                print(f"No data in file: {filename}")
    return project_contributors

# Process OSS projects
oss_project_contributors = process_contributors(oss_dir)

# Process OSS4SG projects
oss4sg_project_contributors = process_contributors(oss4sg_dir)

# Function to build contributor to projects mapping
def build_contributor_projects_map(project_contributors):
    contributor_projects = {}
    for project, contributors in project_contributors.items():
        for contributor in contributors:
            contributor_projects.setdefault(contributor, set()).add(project)
    return contributor_projects

# Build contributor to projects mapping
oss_contributor_projects = build_contributor_projects_map(oss_project_contributors)
oss4sg_contributor_projects = build_contributor_projects_map(oss4sg_project_contributors)

# Calculate the number of projects per contributor
oss_contributors_counts = {contributor: len(projects) for contributor, projects in oss_contributor_projects.items()}
oss4sg_contributors_counts = {contributor: len(projects) for contributor, projects in oss4sg_contributor_projects.items()}

# Convert counts to DataFrames
oss_counts_df = pd.DataFrame(list(oss_contributors_counts.items()), columns=['Contributor', 'Projects Count'])
oss4sg_counts_df = pd.DataFrame(list(oss4sg_contributors_counts.items()), columns=['Contributor', 'Projects Count'])

# Number of contributors in OSS who have contributed to more than two projects
num_oss_contributors_more_than_two = sum(oss_counts_df['Projects Count'] > 1)

# Number of contributors in OSS4SG who have contributed to more than two projects
num_oss4sg_contributors_more_than_two = sum(oss4sg_counts_df['Projects Count'] > 1)

# Average number of projects per contributor in OSS and OSS4SG
average_projects_per_contributor_oss = oss_counts_df['Projects Count'].mean()
average_projects_per_contributor_oss4sg = oss4sg_counts_df['Projects Count'].mean()

# Contributors present in both OSS and OSS4SG
common_contributors = set(oss_contributor_projects.keys()) & set(oss4sg_contributor_projects.keys())
num_common_contributors = len(common_contributors)

# Prepare data for common contributors
common_contributors_df = pd.DataFrame({
    'Contributor': list(common_contributors),
    'OSS Projects Count': [oss_contributors_counts[contributor] for contributor in common_contributors],
    'OSS4SG Projects Count': [oss4sg_contributors_counts[contributor] for contributor in common_contributors]
})

# Visualization

# 1. Bar chart for contributors with more than two projects
fig, ax = plt.subplots(figsize=(10, 6))
counts = [num_oss_contributors_more_than_two, num_oss4sg_contributors_more_than_two]
labels = ['OSS', 'OSS4SG']
ax.bar(labels, counts, color=['skyblue', 'salmon'])
ax.set_title('Number of Contributors with More Than Two Projects')
ax.set_ylabel('Number of Contributors')
for i, v in enumerate(counts):
    ax.text(i, v + 0.5, str(v), ha='center', fontweight='bold')
plt.show()

# 2. Distribution of projects per contributor
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(oss_counts_df['Projects Count'], bins=range(1, oss_counts_df['Projects Count'].max() + 2), ax=axes[0], kde=False, color='skyblue')
axes[0].set_title('OSS Contributors Project Distribution')
axes[0].set_xlabel('Number of Projects')
axes[0].set_ylabel('Number of Contributors')

sns.histplot(oss4sg_counts_df['Projects Count'], bins=range(1, oss4sg_counts_df['Projects Count'].max() + 2), ax=axes[1], kde=False, color='salmon')
axes[1].set_title('OSS4SG Contributors Project Distribution')
axes[1].set_xlabel('Number of Projects')
axes[1].set_ylabel('Number of Contributors')
plt.show()

# 3. Venn diagram of contributors overlap
plt.figure(figsize=(8, 8))
venn2(subsets=(len(oss_contributor_projects) - num_common_contributors,
               len(oss4sg_contributor_projects) - num_common_contributors,
               num_common_contributors),
      set_labels=('OSS Contributors', 'OSS4SG Contributors'),
      set_colors=('skyblue', 'salmon'),
      alpha=0.5)
plt.title('Overlap of Contributors between OSS and OSS4SG')
plt.show()

# 4. Scatter plot for common contributors' project counts
if not common_contributors_df.empty:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=common_contributors_df,
                    x='OSS Projects Count',
                    y='OSS4SG Projects Count',
                    color='green')
    plt.title('Common Contributors Project Counts')
    plt.xlabel('Number of OSS Projects')
    plt.ylabel('Number of OSS4SG Projects')
    plt.show()

# Print statistical summaries
print("Statistical Summary:")
print(f"Total OSS Contributors: {len(oss_contributor_projects)}")
print(f"Total OSS4SG Contributors: {len(oss4sg_contributor_projects)}")
print(f"Number of OSS contributors with >2 projects: {num_oss_contributors_more_than_two}")
print(f"Number of OSS4SG contributors with >2 projects: {num_oss4sg_contributors_more_than_two}")
print(f"Average projects per OSS contributor: {average_projects_per_contributor_oss:.2f}")
print(f"Average projects per OSS4SG contributor: {average_projects_per_contributor_oss4sg:.2f}")
print(f"Number of contributors in both OSS and OSS4SG: {num_common_contributors}")

# Save statistical summaries to a text file
with open(os.path.join(base_dir, 'statistical_summary.txt'), 'w') as f:
    f.write("Statistical Summary:\n")
    f.write(f"Total OSS Contributors: {len(oss_contributor_projects)}\n")
    f.write(f"Total OSS4SG Contributors: {len(oss4sg_contributor_projects)}\n")
    f.write(f"Number of OSS contributors with >1 projects: {num_oss_contributors_more_than_two}\n")
    f.write(f"Number of OSS4SG contributors with >1 projects: {num_oss4sg_contributors_more_than_two}\n")
    f.write(f"Average projects per OSS contributor: {average_projects_per_contributor_oss:.2f}\n")
    f.write(f"Average projects per OSS4SG contributor: {average_projects_per_contributor_oss4sg:.2f}\n")
    f.write(f"Number of contributors in both OSS and OSS4SG: {num_common_contributors}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paths to your CSV files
early_csv = '/Users/mohamadashraf/Downloads/OSS4SG_MSR-main-2/Datasets/GraphInfo/Join Rate vs Leave Rate for early stages.csv'
mid_csv = '/Users/mohamadashraf/Downloads/OSS4SG_MSR-main-2/Datasets/GraphInfo/Join Rate vs Leave Rate for mid stages.csv'
late_csv = '/Users/mohamadashraf/Downloads/OSS4SG_MSR-main-2/Datasets/GraphInfo/Join Rate vs Leave Rate for late stages.csv'

# Read the CSV files
df_early = pd.read_csv(early_csv)
df_mid = pd.read_csv(mid_csv)
df_late = pd.read_csv(late_csv)

# Add 'Stage' column
df_early['Stage'] = 'Early'
df_mid['Stage'] = 'Mid'
df_late['Stage'] = 'Late'

# Combine all dataframes
df_all = pd.concat([df_early, df_mid, df_late], ignore_index=True)

# Set up plotting style
sns.set_style("whitegrid")
sns.set_context("talk")

# Define the median values for dividing the quadrants
join_rate_median = 0.182
leave_rate_median = 0.164

# Split the data into OSS and OSS4SG
df_oss = df_all[df_all['SG'] == 0]  # OSS projects
df_oss4sg = df_all[df_all['SG'] == 1]  # OSS4SG projects

# List of datasets and corresponding titles for OSS and OSS4SG
datasets_oss = [
    ('All Stages Combined - OSS', df_oss),
    ('Early Stage - OSS', df_early[df_early['SG'] == 0]),
    ('Mid Stage - OSS', df_mid[df_mid['SG'] == 0]),
    ('Late Stage - OSS', df_late[df_late['SG'] == 0])
]

datasets_oss4sg = [
    ('All Stages Combined - OSS4SG', df_oss4sg),
    ('Early Stage - OSS4SG', df_early[df_early['SG'] == 1]),
    ('Mid Stage - OSS4SG', df_mid[df_mid['SG'] == 1]),
    ('Late Stage - OSS4SG', df_late[df_late['SG'] == 1])
]

# Function to create the plots
def plot_with_quadrants(datasets, title_prefix, color):
    # Initialize the matplotlib figure with 2x2 subplots
    fig, axs = plt.subplots(2, 2, figsize=(20, 16))
    fig.suptitle(f'{title_prefix} - Join Rate vs Leave Rate Distribution', fontsize=24, fontweight='bold')

    # Flatten the axes array for easy iteration
    axs = axs.flatten()

    # Iterate over each dataset and corresponding subplot axis
    for ax, (title, df) in zip(axs, datasets):
        sns.kdeplot(
            data=df,
            x='Average Leave Rate',
            y='Average Join Rate',
            color=color,  # Set the color based on the type (OSS or OSS4SG)
            fill=True,
            alpha=0.6,
            thresh=0.05,
            levels=10,
            ax=ax
        )

        # Set x and y limits from 0 to 1
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)

        # Customize the title and labels
        ax.set_title(title, fontsize=18, fontweight='bold')
        ax.set_xlabel('Average Leave Rate', fontsize=14, fontweight='bold')
        ax.set_ylabel('Average Join Rate', fontsize=14, fontweight='bold')

        # Add shaded quadrants based on the medians
        ax.axvspan(0, leave_rate_median, ymin=0, ymax=join_rate_median, facecolor='lightgreen', alpha=0.2, label='Stable')
        ax.axvspan(leave_rate_median, 1, ymin=0, ymax=join_rate_median, facecolor='lightcoral', alpha=0.2, label='Unstable')
        ax.axvspan(0, leave_rate_median, ymin=join_rate_median, ymax=1, facecolor='lightblue', alpha=0.2, label='High Leave, Low Join')
        ax.axvspan(leave_rate_median, 1, ymin=join_rate_median, ymax=1, facecolor='lightyellow', alpha=0.2, label='High Leave, High Join')

        # Add vertical and horizontal lines to indicate the median split
        ax.axhline(y=join_rate_median, color='black', linestyle='--', linewidth=1)
        ax.axvline(x=leave_rate_median, color='black', linestyle='--', linewidth=1)

        # Enhance grid
        ax.grid(True, linestyle='--', alpha=0.7)

    # Adjust layout to prevent overlap
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    # Display the plot
    plt.show()

# Plot OSS only (using blue color)
plot_with_quadrants(datasets_oss, "OSS", "blue")

# Plot OSS4SG only (using red color)
plot_with_quadrants(datasets_oss4sg, "OSS4SG", "red")
```
